```python
# get the other hashtags aswell

# regexp for getting hashtags
def other_hashtags(text, main_hashtag):
    #new_text = text.replace("\n",  " ")
    hashtags = re.findall(r'\B#\w*[a-zA-Z]+\w*', text)
    if main_hashtag in hashtags:
        hashtags.remove(main_hashtag)
    return ",".join(hashtags)
```

```python
# get followers for one author
def get_followers(user_id):
    followers = client.get_users_followers(id=user_id)
    #print(followers)
    string_followers = ""
    for user in followers.data:
        string_followers += str(user.id) + ","

    string_followers = string_followers[:-1] # remove last comma
    return string_followers
```

```python
# now get all information from tweets and write in excel file
workbook = xlsxwriter.Workbook('tweets_database.xlsx')
worksheet = workbook.add_worksheet()
row = 0
for tag in range(0, len(tweet_ids)):
    for tweet in range(0, 100):
        print(row)
        tweets_response = client.get_tweet(id=tweet_ids[tag][tweet],expansions=["author_id","geo.place_id"],place_fields="country_code")
        author_id = tweets_response.includes["users"][0].id
        tweet_data = tweets_response.data
        tweet_text = tweet_data.text.lower()
        hashtag = queries[tag]
        included_hashtags = other_hashtags(tweet_text, hashtag)
        try:
            followers = get_followers(author_id)
        except:
            print("ERROR")
            workbook.close()

        retweet = "rt @" in tweet_text #tweet_text.includes("RT @")
        location = ""
        # identify location if included in the tweet
        if 'places' in tweets_response.includes:
            location = tweets_response.includes["places"][0].country_code

        # write data to xlsx file
        worksheet.write(row, 0, hashtag) # main hashtag to first col
        worksheet.write(row, 1, included_hashtags) # second col
        worksheet.write(row, 2, str(author_id))
        worksheet.write(row, 3, tweet_text)
        worksheet.write(row, 4, retweet)
        worksheet.write(row, 5, location)
        worksheet.write(row, 6, followers)
        row = row + 1
    
    
workbook.close()
```
